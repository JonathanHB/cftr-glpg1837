```python
#Lisa uses 1A bin spacing for umbrella sampling and 4 microseconds per bin. Niek's simulations are perpendicular to the membrane
#dynamics are in general 10x faster in coarse grained simulations compared with fine grained ones
# (because their the phase space is equivalent to that of larger particles in size exclusion chromatography)

def sampling_above_threshold(h5path, miniter, maxiter, binbounds):

    max_walker_by_round = []
    pcs_by_bin_round = []
    
    run = Run.open(h5path)

    for iteration in run:
        if iteration.number >= miniter and iteration.number <= maxiter:
            pc_end = np.array([i[-1][-1] for i in iteration.pcoords])
            #samples_by_bin = np.histogram(pc_end, binbounds)[0]
            pcs_by_bin_round.append(pc_end)
            max_walker_by_round.append(np.max(pc_end))

    #print(samples_by_bin_round)

    fig, ax1 = plt.subplots()
    ax1.hist(np.concatenate(pcs_by_bin_round), bins=binbounds)
    plt.ylim(0,500/0.05)

    y_vals = ax1.get_yticks()
    ax1.set_yticklabels(['{:3.0f}'.format(x * 0.05) for x in y_vals])

    plt.xlabel("ligand COM distance from binding site (A)")
    plt.ylabel("nanoseconds of aggregate simulation per Angstrom")

    plt.xlim(0,50)
    
    plt.show()
    
    #a = plt.hist(np.concatenate(pcs_by_bin_round), bins=binbounds)
    #plt.bar(np.sum(np.stack(samples_by_bin_round), axis=0)*0.05)
    # 

    # plt.show()

    plt.plot(max_walker_by_round)
    

we_ind = 3

h5path  = h5paths_names[we_ind][0]
miniter = h5paths_names[we_ind][4]
maxiter = h5paths_names[we_ind][5]

sampling_above_threshold(h5path, miniter, maxiter, binbounds)
```
